```python
from Crypto.Cipher import AES
from Crypto.Util.Padding import pad, unpad
import os
```

```python
#16-byte key and identical plaintext blocks
key= b'Sixteen byte key'
pt= b'IdentityMessage' * 2  #2 identical 16-byte blocks
iv1= os.urandom(16)
iv2= os.urandom(16)
```

```python
def demonstrate_modes():
    print("PART 1: SAME PLAINTEXT TEST")
    
    #ECB MODE
    cipher_ecb= AES.new(key, AES.MODE_ECB)
    ct_ecb= cipher_ecb.encrypt(pad(pt, 16))
    print(f"ECB Ciphertext (Block 1): {ct_ecb[:16].hex()}")
    print(f"ECB Ciphertext (Block 2): {ct_ecb[16:32].hex()}")
    print("Observation: Both blocks are identical in ECB! \n")
    
    #CBC MODE
    cipher_cbc1= AES.new(key, AES.MODE_CBC, iv1)
    ct_cbc1= cipher_cbc1.encrypt(pad(pt, 16))
    cipher_cbc2= AES.new(key, AES.MODE_CBC, iv2)
    ct_cbc2= cipher_cbc2.encrypt(pad(pt, 16))
    
    print(f"CBC (IV 1) Block 1: {ct_cbc1[:16].hex()}")
    print(f"CBC (IV 2) Block 1: {ct_cbc2[:16].hex()}")
    print("Observation: Same plaintext results in different ciphertext with distinct IVs.\n")
    
```
